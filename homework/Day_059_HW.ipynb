### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1], s = 100, c = 'red', label = 'Careful')
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1], s = 100, c = 'blue', label = 'Standard')
plt.scatter(X_pca[y == 2, 0], X_pca[y == 2, 1], s = 100, c = 'green', label = 'Standard')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
pca = PCA(n_components = None)
X_pca = pca.fit_transform(X)

explained_variance = pca.explained_variance_ratio_ 
cumulative_sum = np.zeros_like(explained_variance)
cumulative_sum[0] = explained_variance[0]
for i in range(1 , len(explained_variance)):
    cumulative_sum[i] = cumulative_sum[i - 1] + explained_variance[i]
    
feature_importance = pd.DataFrame({'feature_importance' : explained_variance , 
                                   'cumulative_feature_importance' : cumulative_sum})
feature_importance
```
