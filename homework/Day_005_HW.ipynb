```python
for i , column in enumerate(app_train.columns):
    if app_train[column].dtypes != 'object':
        max_ = app_train[column].max()
        app_train[column].fillna(max_ , inplace = True)
```

```python
app_train['AMT_ANNUITY'].hist(bins = 100)
plt.show()
```

```python
# 轉乘對數型式，整體數據會變得比較像高斯分布

temp = np.clip(app_train['AMT_ANNUITY'] , a_min = 1e-8 , a_max = None)
np.log(temp).hist(bins = 100)
plt.show()
```
